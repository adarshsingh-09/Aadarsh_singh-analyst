```python
#2 Then find the number of days they use their fitness tracker.

user_days = dailyActivity_merged.groupby('Id')['ActivityDate'].nunique().reset_index()
user_days.columns = ['Id', 'TotalDaysUsed']
user_days.head()

# insight 
# The number of days the device was used varies by user, indicating different levels of consistency and engagement with activity tracking
```

```python
#3 Use this data to group all the users into Active(>20), moderate(>10,<20) and
#light(<10) users.

def user_category(days):
    if days > 20:
        return "Active"
    elif 10 <= days <= 20:
        return "Moderate"
    else:
        return "Light"
user_days['UserType'] = user_days['TotalDaysUsed'].apply(user_category)
user_days['UserType'].value_counts()

# insight 
# “Most users fall into the Moderate or Light categories, suggesting inconsistent device usage, while a smaller group shows highly active and regular engagement
```

```python
#4. After that find out the mean distance traveled by each user

mean_distance = dailyActivity_merged.groupby('Id')['TotalDistance'].mean().reset_index()
mean_distance.columns = ['Id', 'MeanDistance']
mean_distance.head()

# insight 
# Average daily distance varies across users, reflecting differences in routine movement patterns and overall fitness levels
```

```python
#5 Use this mean distance to group each user into Pro,Intermediate and
#beginner.(set criteria by yourself)

def distance_level(dist):
    if dist < 5:
        return "Beginner"
    elif dist <= 10:
        return "Intermediate"
    else:
        return "Pro"

mean_distance['FitnessLevel'] = mean_distance['MeanDistance'].apply(distance_level)
mean_distance['FitnessLevel'].value_counts()

# insight 
# Most users fall into the Beginner or Intermediate fitness levels, indicating moderate daily movement, while only a small group shows Pro-level activity
```

```python
#6 Then find out the total steps traveled by each user

total_steps = dailySteps_merged.groupby('Id')['StepTotal'].sum().reset_index()
total_steps.columns = ['Id', 'TotalSteps']
total_steps.head()

# insight 
# Total steps vary significantly among users, highlighting clear differences in overall mobility and daily physical activity levels
```

```python
#7. Then find out the calories burned by each user

calories = pd.read_csv("dailyCalories_merged.csv")
calories_per_user = calories.groupby('Id')['Calories'].sum()
calories_per_user
calories_per_user.head()

# insight 
# Total calories burned differ greatly across users, reflecting varying activity levels, body needs, and consistency in daily movement
```

```python
# 8. And then find out their fairly active, lightly active and Very active minutes.
activity = pd.read_csv("dailyActivity_merged.csv")
active_minutes = activity.groupby('Id')[[
    'FairlyActiveMinutes',
    'LightlyActiveMinutes',
    'VeryActiveMinutes'
]].sum()
active_minutes

# insight 
# Total active minutes vary widely across users, indicating significant differences in daily activity intensity and lifestyle habits
```

```python
#9. Repeat all these tasks (1 to 8) with respect to distinct dates (for example find out
#unique dates instead of unique id and calculate number of users on that date).


daily_summary = activity.groupby('ActivityDate').agg(
    Users=('Id', 'nunique'),
    AvgDistance=('TotalDistance', 'mean'),
    TotalSteps=('TotalSteps', 'sum'),
    Calories=('Calories', 'sum'),
    VeryActive=('VeryActiveMinutes', 'sum'),
    FairlyActive=('FairlyActiveMinutes', 'sum'),
    LightlyActive=('LightlyActiveMinutes', 'sum')
).reset_index()

daily_summary.head()

# insight 
# On each day, overall steps and calories burned rise with higher total active minutes, showing a strong link between user activity level and energy expenditure.
```

```python
# Q11. Balanced Activity Pattern

intensity = intensity.groupby('Id')[[
    'VeryActiveMinutes',
    'FairlyActiveMinutes',
    'LightlyActiveMinutes'
]].std()

intensity.mean(axis=1).sort_values()

# insight 
# Users with lower mean standard deviation show a more consistent activity pattern, while higher values indicate irregular or fluctuating activity levels
```

```python
#Q12. FitWear Premium Customers

days = activity.groupby('Id')['ActivityDate'].nunique()
very_active = intensity.groupby('Id')['VeryActiveMinutes'].mean()

df = pd.concat([days, very_active], axis=1)
df[(df['ActivityDate'] > 20) & (df['VeryActiveMinutes'] > 30)]

# insight 
# Users who logged data for more than 20 days and averaged over 30 very active minutes show a consistently active and disciplined lifestyle
```

```python
#23 High Usage but Low Activity

summary = activity.groupby('Id').agg(
    DaysUsed=('ActivityDate','nunique'),
    AvgSteps=('TotalSteps','mean')
).reset_index()

summary[(summary['DaysUsed'] > 20) & (summary['AvgSteps'] < 5000)]

# insight 
# This output pinpoints users who are consistently engaged but physically inactive
```

```python
# 24 Sedentary Lifestyle Detection

sedentary = activity.groupby('Id')['SedentaryMinutes'].mean().reset_index()
sedentary.sort_values('SedentaryMinutes', ascending=False)

# insight 
# This output highlights who is most inactive on a daily basis
```

```python
#25 Sleep Deprivation Pattern

low_sleep = sleep[sleep['TotalMinutesAsleep'] < 360]
low_sleep.groupby('Id').size().reset_index(name='LowSleepDays') \
         .sort_values('LowSleepDays', ascending=False)


# insight 
# This output highlights users with persistent sleep deprivation patterns, making it a critical insight for health, wellness, and behavior
```

```python
# 26 Sleep Efficiency Analysis


sleep['SleepEfficiency'] = sleep['TotalMinutesAsleep'] / sleep['TotalTimeInBed']

sleep.groupby('Id')['SleepEfficiency'].mean() \
    .sort_values()

# insight 
# Sleep efficiency is often a stronger indicator of sleep quality than total sleep duration.
```

```python
# 27 Calories Burn Stability

cal_var = calories.groupby('Id')['Calories'].std().reset_index()
cal_var.sort_values('Calories')

# insight 
# Higher calorie standard deviation → users show irregular activity intensity, with some very active days and some very inactive days.
```

```python
# 28 Activity Consistency Score

steps_std = activity.groupby('Id')['TotalSteps'].std()
steps_mean = activity.groupby('Id')['TotalSteps'].mean()

consistency = (steps_std / steps_mean).reset_index(name='ConsistencyScore')
consistency.sort_values('ConsistencyScore')

# insight 
# This output ranks users by activity regularity, making it a strong indicator of habit consistency and lifestyle
```

```python
# 29 Light Activity Dependency
activity.groupby('Id')[['LightlyActiveMinutes','VeryActiveMinutes']].mean() \
    .query('LightlyActiveMinutes > 2 * VeryActiveMinutes')


# insight 
# This output identifies users whose routines are dominated by light activity, offering valuable insight into lifestyle balance and fitness improvement opportunities.
```

```python
# 30 Calories per Step Efficiency

eff = activity.groupby('Id').apply(
    lambda x: x['Calories'].sum() / x['TotalSteps'].sum()
).reset_index(name='CaloriesPerStep')

eff.sort_values('CaloriesPerStep', ascending=False)

# insight 
# This output ranks users by calorie burn efficiency per step, highlighting differences in metabolism, body composition, and activity intensity
```

```python
# 32 Data Completeness Check

activity.notnull().groupby(activity['Id']).sum().sum(axis=1) \
    .sort_values(ascending=False)

# insight 
# This output ranks users by data reliability and tracking consistency
```

```python
# 34 Early Warning Inactive Users

activity = activity.sort_values(['Id','ActivityDate'])

trend = activity.groupby('Id')['TotalSteps'].apply(
    lambda x: x.diff().mean()
).reset_index(name='AvgStepChange')

trend.sort_values('AvgStepChange')

# insight 
#This output captures the direction of behavior change, highlighting whether users are becoming more active or less active over time
```

```python
# 35 Sedentary Break Pattern

summary = activity.groupby('Id').agg(
    AvgSteps=('TotalSteps','mean'),
    Sedentary=('SedentaryMinutes','mean')
)

summary[(summary['AvgSteps'] > 8000) & (summary['Sedentary'] > 700)]

# insight
# This output reveals users who are active yet highly sedentary, emphasizing that true lifestyle health depends on movement spread throughout the day, not only total steps.
```

```python
# 36 Sleep Consistency vs Activity Consistency

sleep_var = sleep.groupby('Id')['TotalMinutesAsleep'].std()
step_var = activity.groupby('Id')['TotalSteps'].std()

consistency = pd.concat([sleep_var, step_var], axis=1)
consistency.columns = ['SleepVar','StepVar']

consistency.sort_values('SleepVar')

# insight 
# This output helps evaluate behavioral consistency, highlighting users with stable versus fluctuating sleep and activity patterns
```
