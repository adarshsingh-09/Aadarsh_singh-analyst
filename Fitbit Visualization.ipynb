```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
activity=pd.read_csv('dailyActivity_merged.csv')
activity
```

```python
calories=pd.read_csv('dailyCalories_merged.csv')
calories
```

```python
intensity=pd.read_csv('dailyIntensities_merged.csv')
intensity
```

```python
steps=pd.read_csv('dailySteps_merged.csv')
steps
```

```python
heartrate=pd.read_csv('heartrate_seconds_merged.csv')
heartrate
```

```python
sleep=pd.read_csv('SleepDay_merged.csv')
sleep
```

```python
weight=pd.read_csv('weightlogInfo_merged.csv')
weight
```

```python
# Date column clean & same format 
activity['ActivityDate'] = pd.to_datetime(activity['ActivityDate'])
calories['ActivityDay'] = pd.to_datetime(calories['ActivityDay'])
intensity['ActivityDay'] = pd.to_datetime(intensity['ActivityDay'])
steps['ActivityDay'] = pd.to_datetime(steps['ActivityDay'])
sleep['SleepDay'] = pd.to_datetime(sleep['SleepDay'])
weight['Date'] = pd.to_datetime(weight['Date'])
heartrate['Time'] = pd.to_datetime(heartrate['Time'])
```

```python
# change heartrate 
heartrate['Date'] = heartrate['Time'].dt.date

daily_hr = heartrate.groupby(['Id', 'Date'])['Value'].mean().reset_index()
daily_hr.rename(columns={'Value': 'AvgHeartRate'}, inplace=True)

daily_hr['Date'] = pd.to_datetime(daily_hr['Date'])
```

```python
# merge all dataset
merged_df = (
    activity
    .merge(
        calories,
        left_on=["Id", "ActivityDate"],
        right_on=["Id", "ActivityDay"],
        how="left"
    )
    .merge(
        steps,
        on=["Id", "ActivityDay"],
        how="left"
    )
    .merge(
        intensity,
        on=["Id", "ActivityDay"],
        how="left"
    )
    .merge(
        sleep,
        left_on=["Id", "ActivityDate"],
        right_on=["Id", "SleepDay"],
        how="left"
    )
    .merge(
        weight,
        left_on=["Id", "ActivityDate"],
        right_on=["Id", "Date"],
        how="left"
    )
    .merge(
        daily_hr,
        left_on=["Id", "ActivityDate"],
        right_on=["Id", "Date"],
        how="left"
    )
)
```

```python
# 1 Long-Term Activity Trend (Line Plot)
daily_steps_trend = activity.groupby('ActivityDate')['TotalSteps'].mean()

plt.figure()
plt.plot(daily_steps_trend)
plt.title('Average Daily Steps Over Time')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Average Steps')
plt.show()

# Insight:
# This reveals long-term behavior patterns, consistency, and seasonal spikes or drops in activity.
```

```python
# 2 Activity vs Sedentary Lifestyle (Scatter Plot)

plt.figure()
plt.scatter(activity['TotalSteps'], activity['SedentaryMinutes'])
plt.title('Steps vs Sedentary Minutes')
plt.xlabel('Total Steps')
plt.ylabel('Sedentary Minutes')
plt.show()

# Insight:
# Shows that some users remain sedentary despite high steps, indicating unbalanced daily routines.
```

```python
# 3 Sleep Duration by BMI Category (Bar Chart)

sleep_bmi = sleep.groupby(
    pd.cut(weight['BMI'],
           bins=[0,18.5,25,30,100],
           labels=['Underweight','Normal','Overweight','Obese'])
)['TotalMinutesAsleep'].mean()

plt.figure()
sleep_bmi.plot(kind='bar')
plt.title('Average Sleep Duration by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Minutes Asleep')
plt.show()
# Insight:
#Highlights how physical condition may influence sleep behavior.
```

```python
# 4 Smoothed Activity Pattern (Rolling Average)

rolling_steps = daily_steps_trend.rolling(window=7).mean()

plt.figure()
plt.plot(rolling_steps)
plt.title('7-Day Rolling Average of Steps')
plt.xlabel('Date')
plt.xticks(rotation = 45)
plt.ylabel('Steps')
plt.show()

# Insight:
# Removes noise and highlights the true underlying activity trend
```

```python
# 5 Contribution of Activity Intensities (Stacked Area)

intensity_trend = activity.groupby('ActivityDate')[[
    'VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes'
]].mean()

plt.figure()
plt.stackplot(intensity_trend.index, intensity_trend.T)
plt.title('Activity Intensity Contribution Over Time')
plt.xticks(rotation = 45)
plt.xlabel('Date')
plt.ylabel('Minutes')
plt.show()


# Insight:
# Shows lifestyle composition and dependency on low vs high intensity movement.
```

```python
# 6 Sleep Efficiency Distribution (Histogram)

activity['SleepEfficiency'] = sleep['TotalMinutesAsleep'] / sleep['TotalTimeInBed']

plt.figure()
plt.hist(activity['SleepEfficiency'])
plt.title('Sleep Efficiency Distribution')
plt.xlabel('Sleep Efficiency')
plt.ylabel('Frequency')
plt.show()

# Insight:
# Identifies poor sleep hygiene and inefficient rest patterns.
```

```python
# 7 Calorie Burn by Day of Week (Line Plot)

calories_day = activity.groupby(activity['ActivityDate'].dt.day_name())['Calories'].mean()

plt.figure()
plt.plot(calories_day)
plt.title('Average Calories Burned by Day of Week')
plt.xlabel('Day')
plt.ylabel('Calories')
plt.show()

# Insight:
# Reveals weekday vs weekend lifestyle behavior.
```

```python
# 8 Step Variability Across Users (Bar Plot)

step_variability = activity.groupby('Id')['TotalSteps'].std()

plt.figure(figsize=(10,4))
step_variability.plot(kind='bar')
plt.title('Daily Step Variability per User')
plt.xlabel('User')
plt.ylabel('Step Standard Deviation')
plt.show()

# Insight:
# High variability indicates irregular routines; low variability reflects disciplined habits.
```

```python
# 9 How are key health and activity metrics related to each other when all 7 datasets are merged at the daily level?

heatmap_data = merged_df[[
    'TotalSteps',
    'TotalDistance',
    'Calories_x',
    'VeryActiveMinutes_x',
    'FairlyActiveMinutes_x',
    'LightlyActiveMinutes_x',
    'SedentaryMinutes_x',
    'TotalMinutesAsleep'
]]
corr_matrix = heatmap_data.corr()
plt.figure(figsize=(10,6))
sns.heatmap(
    corr_matrix,
    annot=True,          
    fmt='.2f',
    cmap='YlOrRd',       
    linewidths=0.5,
    linecolor='white'
)
plt.title(
    'Correlation Heatmap of Activity, Sleep and Calorie Metrics',
    fontsize=13,
    weight='bold'
)
plt.tight_layout()
plt.show()
# insight
# This heatmap shows the strength and direction of relationships among key activity, sleep, and calorie variables after merging all datasets.
```

```python
# 10 Sleep Duration Spread (Box Plot)

plt.figure()
plt.boxplot(sleep['TotalMinutesAsleep'].dropna())
plt.title('Sleep Duration Spread')
plt.ylabel('Minutes Asleep')
plt.show()

# Insight:
#Clearly shows typical sleep range and extreme sleepers.
```
